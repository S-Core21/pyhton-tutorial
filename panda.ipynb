NaN - replacing NaN 

```python
I_dict = {"B": "boy", "C": "cat", "D": "dog"}
j_dict = {"C": "cat", "D": "dog", "F": "french"}

I = pd.Series(I_dict)
J = pd.Series(j_dict)
print(J)

k = pd.DataFrame({"I": I, "J": J})
k = {
    "I" : I_dict,
    "J" : j_dict
}
k 
```

```python
k
```

```python
L = k.fillna(10)
L
```

# Try and except in python 
This is python of handling errors gracefully so your program doesnt crash; instead of the program to close or crash because of an error it would help catch the error and decide what to do.

```python
try:
    result = 10 / 0
except:
    print("you cant divide by 0")
```

```python
try:
    result = 10 / 2
    print(result)
except:
    print("you cant divide by 0")
```

```python
try: 
    number = int(input("please input a number"))
    print(100/number)
except ZeroDivisionError:
    print("cannot divide a number by zero")
except ValueError: 
    print("this is not a valid number")
```
